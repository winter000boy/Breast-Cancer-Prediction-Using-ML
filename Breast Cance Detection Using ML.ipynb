## Encoding the Target Variables

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
le = LabelEncoder()
```

### map() function is used to replace the data points

```python
Breast['diagnosis'] = Breast['diagnosis'].map({'M':1, 'B':0})
```

```python
Breast['diagnosis'].value_counts()
```

## Splitting Data Into Training and Testing DataSets

```python
X = Breast.drop('diagnosis',axis =1)  # Beacuase on x axis we want all the other parameters as an input
Y = Breast['diagnosis']  # We want diagnosis as an output on y axix which we are basically trying to predict
```

```python
print(X.shape,Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 42)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

## Standard Scaling (Bringing All the values in a specified Range)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(X_train)
```

```python
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)

X_train
```

## Trainging The Model (Classification Problem) or (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lg = LogisticRegression()
```

```python
lg.fit(X_train,Y_train)     ## X_train is the input and Y_train is the output
prediction = lg.predict(X_test)  ## X_test is the input and we are predicting the output and storing it in prediction variable

prediction
```

```python
str(accuracy_score(Y_test,prediction)*100) + "%"
```
